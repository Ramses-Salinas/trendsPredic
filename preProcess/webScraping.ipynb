```python
from bs4 import BeautifulSoup
import requests
import time

print("Hello")
```

```python
#puesto = input("Puesto: ")
puesto='Desarrollador de Software'
print("Puesto digitado: ",puesto)
#lugar = input("Lugar: ")
lugar = 'Lima, Peru'
print("Lugar digitado: ",lugar)
```

```python
# 1. Obtener el HTML inicial,a partir del cual se sabrá la cantidad de anuncios disponibles
URL_BASE = "https://www.linkedin.com/jobs/search?keywords={}&location={}".format(puesto,lugar)
print("URL: ", URL_BASE)
pedido_obtenido = requests.get(URL_BASE)
html_obtenido = pedido_obtenido.text

print(html_obtenido)
```

```python
# 2. "Parsear" ese HTML
soup = BeautifulSoup(html_obtenido, "html.parser")
#print(soup)
# 3. Obtener el número de ofertas laborales disponibles
count_Jobs = soup.find(attrs={"class": "results-context-header__job-count"})
print(count_Jobs.get_text())

if count_Jobs.get_text() == "1,000+" :
    cantidad=1000
else:
    cantidad=int(count_Jobs.get_text())

print(cantidad)
cuenta = 0
URLs_Jobs = []
print(len(URLs_Jobs))
#cantidad=10
```

```python
#4. Se obtienen los links de las ofertas laborales disponibles
count = int(1)

def obtener_listas(puesto,lugar,index,cuenta,pedido_obtenido):
    URL_BASE_2 = "https://www.linkedin.com/jobs-guest/jobs/api/seeMoreJobPostings/search?keywords={}&location={}&start={}".format(puesto,lugar,index*10)
    print("URL_2: ", URL_BASE_2)
    pedido_obtenido = requests.get(URL_BASE_2)
    html_obtenido = pedido_obtenido.text
    validate = False

    # 2. "Parsear" ese HTML
    soup = BeautifulSoup(html_obtenido, "html.parser")

    # Todas las etiquetas que tengan el atributo "data-tracking-control-name"
    atr_todos = soup.find_all(attrs={"data-tracking-control-name": "public_jobs_jserp-result_search-card"})

    #print(atr_todos)
    type(atr_todos)
    for elemento in atr_todos:
        #if elemento['href'].endswith("public_jobs_jserp-result_search-card"):
        if elemento['href']:
            URLs_Jobs.append(elemento['href'])
            print("_______________________________________________Conseguí: \n",elemento)
            cuenta=cuenta+1
            #if "base-card__full-link" in elemento['class']:
                #print("Conseguí: ",elemento)
            #    cuenta=cuenta+1
            validate = True
            time.sleep(2)
        else:
            validate = False
            time.sleep(2)
            obtener_listas(puesto,lugar,index,cuenta,pedido_obtenido)
    
    if not validate:
        obtener_listas(puesto,lugar,index,cuenta,pedido_obtenido)         

for i in range(0,int(cantidad/10)):
    pedido_obtenido
    obtener_listas(puesto,lugar,i,cuenta,pedido_obtenido)
    print(pedido_obtenido.url)

print(cuenta)
print(pedido_obtenido.url)
print(len(URLs_Jobs))
```
