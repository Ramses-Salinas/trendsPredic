```python
# Define los archivos
archivo_txt = 'JOBs_Content.txt'
df_preConteo = pd.read_csv('conteo_lenguajes_por_año.csv')

print(df_preConteo.head())
```

```python
# Obtener la lista de lenguajes de programación del DataFrame
# Excluir la primera columna 'Año'
lenguajes = df_preConteo.columns[1:]

# Lee el archivo de texto
with open(archivo_txt, 'r', encoding='utf-8') as archivo:
    docs = archivo.readlines()

# Crear un diccionario de textos con identificadores
textos = {f"doc{i+1}": doc.strip() for i, doc in enumerate(docs)}

print(textos)
print(lenguajes)
```

```python
# Función para verificar si una palabra es un lenguaje de programación válido
def es_lenguaje_valido(palabra):
    for lenguaje in lenguajes:
        if palabra == lenguaje.lower():
            return True
```

```python
# Inicializar un contador para contar menciones en los documentos
contador_docs = Counter()
contador_docs_normalizado = Counter()

# Contar menciones de cada lenguaje en los documentos
for texto in textos.values():
    palabras_en_texto = re.split(r'[,:;\s]+', texto.lower())
    #print(palabras_en_texto)
    for palabra in palabras_en_texto:
        for lenguaje in lenguajes:
            if palabra == lenguaje.lower():
                contador_docs[lenguaje] += 1



print(contador_docs)

for lenguaje in lenguajes:
  contador_docs_normalizado[lenguaje] = contador_docs[lenguaje] / sum(contador_docs.values())

print(contador_docs_normalizado)
```

```python
nueva_fila = {lenguaje: (contador_docs_normalizado[lenguaje]) for lenguaje in lenguajes}
# nueva_fila = {}

# Agregar el año correspondiente a la nueva fila
nueva_fila['Año'] = 2024

# Crear una nueva fila con las menciones normalizadas
# for lenguaje in lenguajes:
#    nueva_fila['lenguaje']:contador_docs_normalizado

# Crear un DataFrame con la nueva fila
nueva_fila_df = pd.DataFrame([nueva_fila])

print(nueva_fila)
print(nueva_fila_df)
```

```python
# Asegurarse de que las columnas estén en el mismo orden que el DataFrame original
nueva_fila_df = nueva_fila_df[df_preConteo.columns]

# Concatenar el DataFrame original con la nueva fila
final_df_actualizado = pd.concat([df_preConteo, nueva_fila_df], ignore_index=True)

# Mostrar el DataFrame final
print(final_df_actualizado)
```

```python
# Guardar el DataFrame final en un archivo CSV
final_df_actualizado.to_csv('conteo_lenguajes_por_año_con_docs.csv', index=False)
```
